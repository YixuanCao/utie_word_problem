```python
from sympy import *
import regex as re
x = Symbol('x')
y = Symbol('y')
z = Symbol('z')
unks = [x, y, z]
b = Symbol('b')
cnt = 0
ans_dict = {}
for idx in pairs:
    p = pairs[idx]
    eqs = p[1]
    values = p[2]
    new_eqs = []
    r_ans = p[-1]
    for eq in eqs:
        new_eq = []
        for e in eq:
            if e[0] == 'N':
                e = values[int(e[1:])]
                if '/' in e:
                    new_eq.append('({})'.format(e))
                else:
                    new_eq.append(e)
            else:
                new_eq.append(e)
        new_eq = ''.join(new_eq).split('=')
        new_eq = '{}-({})'.format(new_eq[0], new_eq[1])
        try:
            new_eqs.append(eval(new_eq))
        except:
            print('wrong format of new_eq: ', new_eq, p, data_dict[idx])
    if '|' in r_ans:
        r_ans = r_ans.split('|')[0]
    if 'or' in r_ans:
        r_ans = r_ans.split('or')
        
    try:
        if isinstance(r_ans, list):
            new_r_ans = []
            for ans in r_ans:
                ans = ans.strip('{} ?$').split(';')
                if isinstance(ans, str):
                    ans = [ans]
                for i, a in enumerate(ans):
                    new_r_ans.append((unks[i], round(to_float(re.sub('km\^2|/?[a-zA-Z]|\s', '', a)), 3)))
            r_ans = set(new_r_ans)
        else:
            r_ans = r_ans.strip('{} ?$').split(';')
            r_ans = {(unks[i], round(to_float(re.sub('km\^2|/?[a-zA-Z]|\s', '', a)), 3)) for i, a in enumerate(r_ans)}
    except:
        print('wrong format of real answer:', r_ans,  p, data_dict[idx])
    p_ans = solve(new_eqs)
    try:
        if isinstance(p_ans, list):
            new_p_ans = []
            for ans in p_ans:
                for k, v in ans.items():
                    new_p_ans.append((k, round(float(v), 3)))
            p_ans = set(new_p_ans)
        else:
            p_ans = {(k, round(float(v), 3)) for k, v in p_ans.items()}
    except:
        print('wrong format of predict answer: ', type(p_ans), p_ans,  p, data_dict[idx])
    if isinstance(p_ans, set) and isinstance(r_ans, set) and r_ans.issubset(p_ans):
        cnt += 1
    else:
        print(new_eqs)
        print('r_ans, p_ans: ', r_ans, p_ans, p, data_dict[idx])
    ans_dict[idx] = p_ans
```

```python
equations = []
def replace_eq(op, rel2eq):
    # 从长找到短
    for rel in rel2eq:
        if rel[0] in op:
            op = op.replace(rel[0], rel[1])
            replace_eq(op, rel2eq)
    else:
        return op
    
def replace_v(op, wid2value):
    for wid, v in wid2value.items():
        if wid in op:
            op = op.replace(wid, v)
            replace_v(op, wid2value)
    else:
        return op
eq_ans = []
for sam in avail_results:
    rel2eq = {}
    eqs = []
    _, _, values, _, ans = sam['info']['q']
    idx = 0
    wid2value = {}
    ent2v = {}
    for i, w in enumerate(sam['words']):
        if w['word'] == 'n' and not sam['words'][i+1]['word'].startswith('##'):
            wid2value[w['id']] = values[idx]
            idx += 1
        else:
            wid2value[w['id']] = w['word']
    for e in sam['entities']:
        if e['tokens'] in wid2value:
            if '/' in wid2value[e['tokens']]:
                ent2v[e['id']] = '({})'.format(wid2value[e['tokens']])
            else:
                ent2v[e['id']] = wid2value[e['tokens']]
    for rel in sam['relations']:
        if rel['type'] != '=':
            rel2eq[rel['id']] = '({}{}{})'.format(rel['operands'][0], rel['type'], rel['operands'][1])
        if rel['type'] == '=':
            eqs.append('{}-{}'.format(rel['operands'][0], rel['operands'][1]))
    rel2eq = sorted(rel2eq.items(), key=lambda x: len(x[0]), reverse=True)
    new_eqs = []
    for eq in eqs:
        eq = replace_eq(eq, rel2eq)
        eq = replace_v(eq, ent2v)
        new_eqs.append(eval(eq))
    eq_ans.append((new_eqs, ans))
    print('origin_data: ',sam['info']['q'], sam['relations'])
    print('new_eqs: ', new_eqs)
```

```python
omit1 = re.compile('(\d+)([a-z])')
omit2 = re.compile('(\d+|[a-z])(\()')
omit3 = re.compile('(\))(\d+|[a-z])') 
omit4 = re.compile('(\))(\()')                                    
```

```python
def add_mul_in_eq(text):
    # 把 12,345 变成 12345
    def comma_f(m):
        groups = m.groups()
        return '{}*{}'.format(groups[0], groups[1])
    text = omit1.sub(repl=comma_f, string=text)
    text = omit2.sub(repl=comma_f, string=text)
    text = omit3.sub(repl=comma_f, string=text)
    text = omit4.sub(repl=comma_f, string=text)
    return text
```
